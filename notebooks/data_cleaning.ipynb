Data Cleaning

In this section, we clean the COVID-19 vaccination dataset to prepare it for analysis. Steps include:

- Handling missing numeric values (forward fill for cumulative columns)
- Converting the date column to datetime
- Removing duplicates
- Standardizing column names
- Handling categorical data in the vaccines column
- Filtering out invalid or negative values

```python
# identifying missing values
df.isnull().sum()
```

```python
# Check number and percentage of missing values
missing_count = df.isnull().sum()
missing_percent = df.isnull().mean() * 100

missing_df = pd.DataFrame({
    'missing_count': missing_count,
    'missing_percent': missing_percent
}).sort_values(by='missing_percent', ascending=False)

missing_df
```

Handling Missing Values
- Cumulative columns (total_vaccinations, people_vaccinated, etc.): forward-fill per country to propagate last known values, then fill any starting NaNs with 0.
- Daily columns (daily_vaccinations, daily_vaccinations_raw): fill missing values with 0 to represent days with no reported vaccinations.
- Other columns like country, date, iso_code are fully populated and require no cleaning.

```python
# Cumulative columns: forward-fill then fill starting NaNs with 0
cumulative_cols = [
    'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
    'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
    'people_fully_vaccinated_per_hundred'
]
df[cumulative_cols] = df.groupby('country')[cumulative_cols].ffill().fillna(0)
```

```python
# Daily columns: fill missing with 0
daily_cols = ['daily_vaccinations_raw', 'daily_vaccinations', 'daily_vaccinations_per_million']
df[daily_cols] = df[daily_cols].fillna(0)
```

```python
# Converting the date column to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Verify missing values
df.isnull().sum()
```

```python
# standardize column names
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
```

```python
# remove duplicates 
df = df.drop_duplicates()
```

```python
df['vaccines'] = df['vaccines'].str.replace(', ', ',')  # normalize separators
```

```python
df['vaccine_list'] = df['vaccines'].str.split(',')
```

```python
#remove negative numbers
df = df[df['daily_vaccinations'] >= 0]
```

```python
# save cleaned dataset
df.to_csv("../data/processed/country_vaccinations_clean.csv", index=False)
```
